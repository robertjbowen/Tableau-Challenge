```python
# https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input=Museum%20of%20Contemporary%20Art%20Australia&inputtype=textquery&fields=photos,formatted_address,name,rating,opening_hours,geometry&key=YOUR_API_KEY
# https://maps.googleapis.com/maps/api/place/nearbysearch/json?location=-33.8670522,151.1957362&radius=1500&type=restaurant&keyword=cruise&key=YOUR_API_KEY
#Create empty list to hold all of the hotel data
zip_list = []

#Format API request
stations = zip_citi['start station name']
#stations = ['E 16 St & Irving Pl', 'West St & Chambers St']
#url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"
url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json?"

#Print header for API processing log
print('Beginning Data Retrieval ') 
print('-----------------------------')

#Conduct a series of successive API calls to gather zipcode data for vacation cities
for station in stations: 
    print(f'Retrieving Results for Station | {station}') 
    df = zip_citi[zip_citi['start station name'] == station] 
    station_lat = df.iloc[0]['start station latitude'] 
    station_lng = df.iloc[0]['start station longitude'] 
    # parameters = (f"location={station_lat},{station_lng}&radius=50&key={g_key}")
    parameters = (f"input=Citi%20Bike&inputtype=textquery&locationbias=point:{station_lat},{station_lng}&fields=formatted_address&key={g_key}")

#Check to make sure the city has hotel data, if not skip that city and continue
    try:

#Get the response from the API request and capture the hotel data values from the request
        response = requests.get(url + parameters).json()  ###Clear this to run the api query###
        print(response)
#extract results
        address = response['candidates'][0]['formatted_address']
        print(f"Address is {address}.")
        arr = address.split(", ")
        zippy = arr[2].split(" ")
        zip = zippy[1]
        zip_list.append(zip)

#Generate an exception message if hotel data not found
    except Exception:
        print(f'{station} station not found. Skipping...')
        
zip_citi['zip code'] = zip_list
zip_citi
#zip_list
```
